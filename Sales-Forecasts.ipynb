```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
import scipy.stats as stats

# Assuming df is already loaded
# Filtering the data for the 'Phones' sub-category
df_phones = df[df['Sub-Category'] == 'Phones'].copy()

# Grouping by order date for monthly sales
df_phones['Order Date'] = pd.to_datetime(df_phones['Order Date'])
df_phones.set_index('Order Date', inplace=True)
monthly_sales = df_phones['Sales'].resample('ME').sum()

# Differencing to make the data stationary if necessary
diff_sales = monthly_sales.diff().dropna()

# Fit SARIMA model (you can adjust p, d, q, P, D, Q, m parameters based on analysis)
model = SARIMAX(diff_sales, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12), enforce_stationarity=False, enforce_invertibility=False)
results = model.fit(disp=False)

# Check AIC and BIC
aic = results.aic
bic = results.bic
print(f"AIC: {aic}, BIC: {bic}")

# Residual Plot
residuals = results.resid
plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title("Residuals of SARIMA Model")
plt.show()

# Residual Diagnostics
# 1. ACF and PACF of Residuals
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(residuals, ax=ax[0])
ax[0].set_title("ACF of Residuals")
plot_pacf(residuals, ax=ax[1])
ax[1].set_title("PACF of Residuals")
plt.show()

# 2. Q-Q Plot
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals")
plt.show()

# 3. Variance Check
print(f"Variance of Residuals: {np.var(residuals)}")

# 4. Ljung-Box Test
ljung_box_result = acorr_ljungbox(residuals, lags=[10], return_df=True)
print(ljung_box_result)

# If residuals are sufficiently good, perform forecasting
if ljung_box_result['lb_pvalue'].iloc[0] > 0.05:
    forecast = results.get_forecast(steps=12)
    forecast_ci = forecast.conf_int()

    # Plotting the forecast
    plt.figure(figsize=(12, 6))
    plt.plot(diff_sales, label='Observed')
    plt.plot(forecast.predicted_mean, label='Forecast', color='r')
    plt.fill_between(forecast_ci.index,
                     forecast_ci.iloc[:, 0],
                     forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
    plt.title('Forecast for Phones Sales')
    plt.legend()
    plt.show()
else:
    print("Model residuals are not sufficiently good. Consider model re-specification or transformations.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import boxcox  # Import Box-Cox transformation
from scipy.special import inv_boxcox  # Import inverse Box-Cox transformation

# Assuming df is already loaded
# Filtering the data for the 'Phones' sub-category
df_phones = df[df['Sub-Category'] == 'Phones'].copy()

# Grouping by order date for monthly sales
df_phones['Order Date'] = pd.to_datetime(df_phones['Order Date'])
df_phones.set_index('Order Date', inplace=True)
monthly_sales = df_phones['Sales'].resample('M').sum()

# Apply Box-Cox transformation to stabilize variance
# Box-Cox requires positive data; shift data if necessary
if (monthly_sales <= 0).any():
    shift = abs(monthly_sales.min()) + 1
    monthly_sales_shifted = monthly_sales + shift
    sales_transformed, lam = boxcox(monthly_sales_shifted)
else:
    shift = 0
    sales_transformed, lam = boxcox(monthly_sales)
print(f"Optimal Lambda for Box-Cox Transformation: {lam}")

# Split the data into train and test sets
train_size = int(len(sales_transformed) * 0.8)
train_transformed, test_transformed = sales_transformed[:train_size], sales_transformed[train_size:]

# Perform Grid Search using auto_arima to find the best SARIMA model
model = auto_arima(train_transformed, seasonal=True, m=12, stepwise=True, trace=True, suppress_warnings=True)

# Print model summary
print(model.summary())

# Fit the best model (auto_arima returns a fitted model)
results = model

# Check AIC and BIC
aic = results.aic()
bic = results.bic()
print(f"AIC: {aic}, BIC: {bic}")

# Residual Plot
residuals = results.resid()
plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title("Residuals of SARIMA Model (Box-Cox Transformed Data)")
plt.show()

# Residual Diagnostics
# 1. ACF and PACF of Residuals
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(residuals, ax=ax[0])
ax[0].set_title("ACF of Residuals")
plot_pacf(residuals, ax=ax[1])
ax[1].set_title("PACF of Residuals")
plt.show()

# 2. Q-Q Plot
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals")
plt.show()

# 3. Variance Check
print(f"Variance of Residuals: {np.var(residuals)}")

# 4. Ljung-Box Test
ljung_box_result = acorr_ljungbox(residuals, lags=[10], return_df=True)
print(f"Ljung-Box Test Results:\n{ljung_box_result}\n")

# If residuals are sufficiently good, perform forecasting
if ljung_box_result['lb_pvalue'].iloc[0] > 0.05:
    # Forecasting using pmdarima's predict method
    forecast_transformed, conf_int_transformed = results.predict(n_periods=len(test_transformed), return_conf_int=True)

    # Inverse Box-Cox transformation to get back to original scale
    if shift != 0:
        # Reverse the shift after inverse transformation
        forecast = inv_boxcox(forecast_transformed, lam) - shift
        conf_int = inv_boxcox(conf_int_transformed, lam) - shift
    else:
        forecast = inv_boxcox(forecast_transformed, lam)
        conf_int = inv_boxcox(conf_int_transformed, lam)

    # Align forecast index with test data index
    forecast_index = monthly_sales.index[train_size:]

    # Actual test data on original scale
    test_actual = monthly_sales[train_size:]

    # Plotting the forecast vs test data
    plt.figure(figsize=(12, 6))
    plt.plot(monthly_sales.index[:train_size], monthly_sales[:train_size], label='Training Data')
    plt.plot(test_actual.index, test_actual, label='Actual Sales', color='blue')
    plt.plot(forecast_index, forecast, label='Forecast', color='red')
    plt.fill_between(forecast_index,
                     conf_int[:, 0],
                     conf_int[:, 1], color='pink', alpha=0.3)
    plt.title('Forecast vs Actual for Phones Sales (Box-Cox Transformed Model)')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend()
    plt.show()

    # Calculate forecast accuracy
    mse = mean_squared_error(test_actual.values, forecast)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(test_actual.values, forecast)
    mape = np.mean(np.abs((test_actual.values - forecast) / test_actual.values)) * 100
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
else:
    print("Model residuals are not sufficiently good. Consider model re-specification or transformations.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy.stats import probplot
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df is already loaded
# Filtering the data for the 'Phones' sub-category
df_phones = df[df['Sub-Category'] == 'Phones'].copy()

# Grouping by order date for monthly sales
df_phones['Order Date'] = pd.to_datetime(df_phones['Order Date'])
df_phones.set_index('Order Date', inplace=True)
monthly_sales = df_phones['Sales'].resample('M').sum()

# Split the data into train and test sets
train_size = int(len(monthly_sales) * 0.8)
train, test = monthly_sales[:train_size], monthly_sales[train_size:]

# 1) Apply SARIMA with manually inputted parameters
sarima_order = (1, 1, 1)
sarima_seasonal_order = (1, 1, 1, 12)
sarima_model = SARIMAX(train, order=sarima_order, seasonal_order=sarima_seasonal_order, enforce_stationarity=False, enforce_invertibility=False)

# 2) Fit the SARIMA model and print summary
sarima_results = sarima_model.fit(disp=False)
print(sarima_results.summary())

# 3) Residual diagnostics
residuals = sarima_results.resid

# Q-Q Plot
plt.figure(figsize=(10, 6))
probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of SARIMA Residuals")
plt.show()

# Ljung-Box Test
ljung_box_result = acorr_ljungbox(residuals, lags=[10], return_df=True)
print("Ljung-Box Test:")
print(ljung_box_result)

# ACF and PACF of Residuals
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(residuals, ax=ax[0])
ax[0].set_title("ACF of Residuals")
plot_pacf(residuals, ax=ax[1])
ax[1].set_title("PACF of Residuals")
plt.show()

# 4) SARIMA Forecast vs Real Values
sarima_forecast = sarima_results.get_forecast(steps=len(test))
sarima_forecast_mean = sarima_forecast.predicted_mean
sarima_forecast_ci = sarima_forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train', color='blue')
plt.plot(test.index, test, label='Test', color='orange')
plt.plot(sarima_forecast_mean.index, sarima_forecast_mean, label='SARIMA Forecast', color='red')
plt.fill_between(sarima_forecast_mean.index,
                 sarima_forecast_ci.iloc[:, 0],
                 sarima_forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.title('SARIMA Forecast vs Real Values')
plt.legend()
plt.show()

# 5) Forecast using Facebook Prophet
prophet_df = monthly_sales.reset_index()
prophet_df.columns = ['ds', 'y']

# Split the data into train and test sets
train_prophet, test_prophet = prophet_df[:train_size], prophet_df[train_size:]

# Define and fit the Facebook Prophet model
prophet_model = Prophet()
prophet_model.fit(train_prophet)

# Make future dataframe and predict
future = prophet_model.make_future_dataframe(periods=len(test_prophet), freq='ME')
forecast_prophet = prophet_model.predict(future)

# Plotting just the Prophet prediction
plt.figure(figsize=(12, 6))
plt.plot(test_prophet['ds'], test_prophet['y'], label='Test', color='orange')
plt.plot(forecast_prophet['ds'], forecast_prophet['yhat'], label='Prophet Forecast', color='green')
plt.title('Prophet Forecast vs Real Values')
plt.legend()
plt.show()

# 6) Calculate errors for SARIMA and Prophet models
# SARIMA Errors
sarima_mae = mean_absolute_error(test, sarima_forecast_mean)
sarima_rmse = np.sqrt(mean_squared_error(test, sarima_forecast_mean))
sarima_mape = np.mean(np.abs((test - sarima_forecast_mean) / test)) * 100
print(f"SARIMA Mean Absolute Error (MAE): {sarima_mae}")
print(f"SARIMA Root Mean Squared Error (RMSE): {sarima_rmse}")
print(f"SARIMA Mean Absolute Percentage Error (MAPE): {sarima_mape}%")

# Prophet Errors
prophet_forecast_test = forecast_prophet[['ds', 'yhat']].set_index('ds').loc[test_prophet['ds']]
prophet_mae = mean_absolute_error(test_prophet['y'], prophet_forecast_test['yhat'])
prophet_rmse = np.sqrt(mean_squared_error(test_prophet['y'], prophet_forecast_test['yhat']))
prophet_mape = np.mean(np.abs((test_prophet['y'] - prophet_forecast_test['yhat']) / test_prophet['y'])) * 100
print(f"Prophet Mean Absolute Error (MAE): {prophet_mae}")
print(f"Prophet Root Mean Squared Error (RMSE): {prophet_rmse}")
print(f"Prophet Mean Absolute Percentage Error (MAPE): {prophet_mape}%")
```
